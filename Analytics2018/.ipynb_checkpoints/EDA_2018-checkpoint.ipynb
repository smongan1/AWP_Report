```python
def resize(text, length_limit = 50):
    text = text.split(' ')
    out = ['']
    for x in text:
        if (len(x) + len(out[-1]) + 1) >= length_limit:
            out.append('')
        if out[-1]:
            out[-1] += ' ' + x
        else:
            out[-1] = x
    out = '\n'.join(out)
    return out
```

```python
with open("Encodings.dat", 'rb') as f:
    conversion_dictionary = pickle.load(f)

def encode(df, encoding_dictionary):
    if isinstance(df, pd.DataFrame):
        df = df.copy()
        for x in df:
            df[x] = df[x].apply(encode, args = [encoding_dictionary])
        return df
    if df in encoding_dictionary:
        df = encoding_dictionary[df]
    return df
```

```python
df = df_orig.copy()
df_onehot = pd.get_dummies(df[[x for x in df if types[x] == 'qualitative']])
df = pd.concat([df, df_onehot], axis = 1)
df = df.drop(columns = r_types['nan/drop'] + r_types['qualitative'])
df[r_types['quantitative']] = encode(df[r_types['quantitative']], conversion_dictionary)

for col in r_types['quantitative']:
    un = df[col].unique()
    if len(un) == 2:
        print(col)
        df[col] = df[col].isna()
    if len(un) == 1:
        df = df.drop(columns = col)
df['description'] = ''
for col in r_types['description']:
    df['description'] = df['description'] + ' ' + df[col].fillna(' ')
df = df.drop(columns = r_types['description'])
```

```python
feet_col = 'Feet:About how tall are you without shoes? Please estimate to the best of your ability.'
df[feet_col] = df[feet_col].apply(lambda x : np.nan if isinstance(x, str) else x).astype('float')
inch_col = 'Inches:About how tall are you without shoes? Please estimate to the best of your ability.'
df['height_inches'] = (12 * df[feet_col] + 
df[inch_col].apply(lambda x: np.nan if isinstance(x, str) else x).astype('float'))
df['LengthOfService'] = (df['What year did your status become retired/discharged? (Ex: 2006)']
                         -df['What year did you begin your time in the service? (Ex: 1985)'])
```

```python
dfPredict = df.copy()
dropCols = []
for col in dfPredict:
    dfPredict[col] = dfPredict[col].fillna(dfPredict[col].mode())
    try: 
        dfPredict[col] = dfPredict[col].fillna(dfPredict[col].mean())
        dfPredict[col] = dfPredict[col].astype(float)
    except: 
        dropCols.append(col)
dfPredict = dfPredict.drop(columns = dropCols)
dfPredict = dfPredict.drop(columns = ['What year did your status become retired/discharged? (Ex: 2006)',
               'What year did you begin your time in the service? (Ex: 1985)'])

dfPredict = dfPredict.drop(columns = [x for x in dfPredict if 'feet:' in x.lower() or 'inches:' in x.lower()])
```
